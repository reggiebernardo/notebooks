### PTA 2p-function/statistics analysis-demo notebook

In this notebook, present the computation of the two-point function statistics in PTA simulations; utilizing our `pulsar_2p_stats` module.

Then, we import pre-computed samples of two-point statistics of pulsar timing residuals in mock pulsar timing array (PTA) simulations, and analyze the results using our module `pulsar_2p_class` to visualize the Hellings-Downs (HD) curve and the higher cumulants of the two-point function of a gravitational wave background (GWB). See [arXiv:2407.17987](https://arxiv.org/abs/2407.17987) for more details.

```python
%%capture
import numpy as np
import matplotlib.pyplot as plt

from pulsar_2p_stats import *
from pulsar_2p_class import *
```

#### 0 PTA two-point statistics using `pulsar_2p_stats`

To compute the two-point function of a GWB in PTA, we first bin the pulsars in angular separation; afterwards we bin each pair in each angular bin in frequency bins. The actual steps of the computation is carried out in the `pulsar_2p_stats` module. We demonstrate the use of this module using the master function `ppkstats_master` below by computing the two-point statistics for noise-free and mild-noise cases with 50 pulsars and 30 realisations.

(Note that this computation takes a while due to plenty pulsar pairs and realisations; so take a coffee or something.)

```python
# fixed input parameters
npsrs=50
Tspan_yr=30

# gwb, noise input
add_gwb=True
gamma_gw=13/3; log10_A_gw=np.log10(2.4e-15)

# some defaults, change if needed
ntoas=100; toaerr=1e-7

# methodology input
n_bins=14 # number of frequency bins
na_bins=15 # number of angular bins
n_sims=30 # number of realisations

# variable inputs, rn_input_dict
rn_input_dict_mild={'log10_A_min':-16, 'log10_A_max': -14, 'gamma_min': 2, 'gamma_max': 6}

free=ppkstats_master(npsrs=npsrs, Tspan_yr=Tspan_yr, ntoas=ntoas, toaerr=toaerr, \
                     add_gwb=add_gwb, gamma_gw=gamma_gw, log10_A_gw=log10_A_gw, \
                     add_red=False, add_white=False, \
                     n_bins=n_bins, na_bins=na_bins, n_sims=n_sims, \
                     save_pickle=True, pkl_name='noise_free', show_progress=True)

# mild noise (rn + gwb)
mild=ppkstats_master(npsrs=npsrs, Tspan_yr=Tspan_yr, ntoas=ntoas, toaerr=toaerr, \
                     add_gwb=add_gwb, gamma_gw=gamma_gw, log10_A_gw=log10_A_gw, \
                     add_red=True, rn_input_dict=rn_input_dict_mild, add_white=False, \
                     n_bins=n_bins, na_bins=na_bins, n_sims=n_sims, \
                     save_pickle=True, pkl_name='mild_noise', show_progress=True)
```

The output can be analyzed by loading the data using the `Pulsar2pStats` class as follows.

```python
# check /data for pre-computed samples by ppkstats_master
npsrs=50
n_sims=30
free=Pulsar2pStats(name='noise_free', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=0)
free.load_data(print_input=True);

mild=Pulsar2pStats(name='mild_noise', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=1)
mild.load_data(print_input=True);
```

The PTA two-point function statistics in the 14th frequency bin for the noise-free and mild-noise cases that were computed are now shown.

```python
k_index=13 # (k_index+1)th-frequency bin
ra2_scaled=(4e-12)*(k_index+1)**(-13/3) # rescale amplitude according to GWB power spectrum
ppkstats=free.ppkstats
fig, ax=plt.subplots(nrows=4,ncols=2,sharex=True, sharey='row', figsize=(10,10))
free.plot_rkrk(ax, k_index=k_index, color='blue', fmt='o', markersize=5, elinewidth=1, capsize=5)
free.plot_rkrk_model(ax, ra2=ra2_scaled, color='blue', ls='-')
mild.plot_rkrk(ax, k_index=k_index, color='red', fmt='s', markersize=5, elinewidth=1, capsize=1)
    
ax[0,0].set_ylabel(r'${\cal E}_1$ ($\mu$s$^2$)')
ax[1,0].set_ylabel(r'${\cal V}_2$ ($\mu$s$^4$)')
ax[2,0].set_ylabel(r'${\cal S}_3$ ($\mu$s$^6$)')
ax[3,0].set_ylabel(r'${\cal K}_4$ ($\mu$s$^8$)')
ax[0,0].set_title(rf'$\alpha_{{(a{k_index+1})}} \alpha_{{(b{k_index+1})}}$')
ax[0,1].set_title(rf'$\beta_{{(a{k_index+1})}} \beta_{{(b{k_index+1})}}$')

ax[1,0].set_yscale('log'); ax[1,1].set_yscale('log')
ax[2,0].set_yscale('symlog'); ax[2,1].set_yscale('symlog')
ax[3,0].set_yscale('log'); ax[3,1].set_yscale('log')

ax[3,0].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,1].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,0].set_xticks(np.arange(0,180,20))
ax[3,0].set_xlim(0, 180); ax[3,1].set_xlim(0, 180)

# move ticks to right for right column axes
for row in range(len(ax)):
    ax[row, 1].yaxis.tick_right()
    ax[row, 1].tick_params(labelright=True, labelleft=False)
fig.tight_layout()
fig.subplots_adjust(hspace=0,wspace=0)
plt.show()
```

#### 1 Two-point statistics of a nanohertz GWB

We import pre-computed samples (100 pulsars, 300 realisations each case) of two-point statistics in mock pulsar timing array (PTA) simulations, and analyze the cumulants of the PTA/GWB two-point function. This recreates the results in [arXiv:2407.17987](https://arxiv.org/abs/2407.17987).

We load the data using the `Pulsar2pStats` class. We consider several cases (noise-free, soft-, mild-, and loud-noise) depending on the red noise loudness compared with GWB; values can be inspected in the printed input or by loading the pickle files directly.

```python
# check /data for pre-computed samples
npsrs=100 # available: 100, 200, 400
n_sims=300 # available: 300, 5000 (for 100 pulsars only)
free=Pulsar2pStats(name='noise_free', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=0)
free.load_data(print_input=True);

soft=Pulsar2pStats(name='soft_noise', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=1)
soft.load_data(print_input=True);

mild=Pulsar2pStats(name='mild_noise', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=1)
mild.load_data(print_input=True);

loud=Pulsar2pStats(name='loud_noise', npsrs=npsrs, n_sims=n_sims, gwb_binary=1, rn_binary=1)
loud.load_data(print_input=True);
```

We start with the noise-free case or the GWB signal. Plot below shows the mean (HD curve) and the higher cumulants of the GWB two-point function for a nanohertz GWB; note that the data is binned in both frequency and angular separation between pulsars to construct the two-point function.

```python
k_index=0
ra2_scaled=(4e-12)*(k_index+1)**(-13/3) # rescale amplitude according to GWB power spectrum
ppkstats=free.ppkstats
fig, ax=plt.subplots(nrows=4,ncols=2,sharex=True, sharey='row', figsize=(10,10))
free.plot_rkrk(ax, k_index=k_index, color='blue', fmt='o', markersize=5, elinewidth=1, capsize=5)
free.plot_rkrk_model(ax, ra2=ra2_scaled, color='blue', ls='-')
    
ax[0,0].set_ylabel(r'${\cal E}_1$ ($\mu$s$^2$)')
ax[1,0].set_ylabel(r'${\cal V}_2$ ($\mu$s$^4$)')
ax[2,0].set_ylabel(r'${\cal S}_3$ ($\mu$s$^6$)')
ax[3,0].set_ylabel(r'${\cal K}_4$ ($\mu$s$^8$)')
ax[0,0].set_title(rf'$\alpha_{{(a{k_index+1})}} \alpha_{{(b{k_index+1})}}$')
ax[0,1].set_title(rf'$\beta_{{(a{k_index+1})}} \beta_{{(b{k_index+1})}}$')

ax[1,0].set_yscale('log'); ax[1,1].set_yscale('log')
ax[3,0].set_yscale('log'); ax[3,1].set_yscale('log')

ax[3,0].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,1].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,0].set_xticks(np.arange(0,180,20))
ax[3,0].set_xlim(0, 180)
ax[3,1].set_xlim(0, 180)

# move ticks to right for right column axes
for row in range(len(ax)):
    ax[row, 1].yaxis.tick_right()
    ax[row, 1].tick_params(labelright=True, labelleft=False)
fig.tight_layout()
fig.subplots_adjust(hspace=0,wspace=0)
plt.show()
```

The signatures of Gaussianity are in the higher cumulants of the two-point function; the mean is the HD curve and its cosmic variance, while the higher cumulants are functions of the HD curve due to Gaussianity.

The above are obtained from PTA residuals, without the noise components. The resulting two-point statistics with noise are shown in the following; soft---red noise (RN) up to the same order of magnitude in amplitude compared with GWB; mild---up to an order of magnitude above GWB; load---RN up to two orders of magnitude above GWB.

```python
k_index=0
ra2_scaled=(4e-12)*(k_index+1)**(-13/3) # rescale amplitude according to GWB power spectrum
ppkstats=free.ppkstats
fig, ax=plt.subplots(nrows=4,ncols=2,sharex=True, sharey='row', figsize=(10,10))
free.plot_rkrk(ax, k_index=k_index, color='blue', fmt='o', markersize=5, elinewidth=1, capsize=5)
free.plot_rkrk_model(ax, ra2=ra2_scaled, color='blue', ls='-')

soft.plot_rkrk(ax, k_index=k_index, color='green', fmt='*', markersize=5, elinewidth=1, capsize=3)
mild.plot_rkrk(ax, k_index=k_index, color='red', fmt='s', markersize=5, elinewidth=1, capsize=1)
# loud.plot_rkrk(ax, k_index=k_index, color='black', fmt='^', markersize=1, elinewidth=1, capsize=0)
    
ax[0,0].set_ylabel(r'${\cal E}_1$ ($\mu$s$^2$)')
ax[1,0].set_ylabel(r'${\cal V}_2$ ($\mu$s$^4$)')
ax[2,0].set_ylabel(r'${\cal S}_3$ ($\mu$s$^6$)')
ax[3,0].set_ylabel(r'${\cal K}_4$ ($\mu$s$^8$)')
ax[0,0].set_title(rf'$\alpha_{{(a{k_index+1})}} \alpha_{{(b{k_index+1})}}$')
ax[0,1].set_title(rf'$\beta_{{(a{k_index+1})}} \beta_{{(b{k_index+1})}}$')

ax[0,0].set_yscale('symlog'); ax[0,1].set_yscale('symlog')
ax[1,0].set_yscale('log'); ax[1,1].set_yscale('log')
ax[2,0].set_yscale('symlog'); ax[2,1].set_yscale('symlog')
ax[3,0].set_yscale('log'); ax[3,1].set_yscale('log')

ax[2,0].set_yticks([-1e8,-1e4,0,1e4,1e8]); ax[2,1].set_yticks([-1e8,-1e4,0,1e4,1e8])

ax[3,0].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,1].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,0].set_xticks(np.arange(0,180,20))
ax[3,0].set_xlim(0, 180); ax[3,1].set_xlim(0, 180)
# move ticks to right for right column axes
for row in range(len(ax)):
    ax[row, 1].yaxis.tick_right()
    ax[row, 1].tick_params(labelright=True, labelleft=False)
fig.tight_layout()
fig.subplots_adjust(hspace=0,wspace=0)
plt.show()
```

The above plot shows that despite being uncorrelated the RN dominates the higher cumulants of the two-point function in PTA, particularly the variance and the kurtosis. An explanation to this is given in [2407.17987](https://arxiv.org/abs/2407.17987).

One can see nonetheless that for both the soft and mild noise cases that the signal is already manifest in the mean, each one resembling the HD curve and its cosmic variance. It only remains to see the signatures of Gaussianity in the higher cumulants. This is shown in the higher frequency bins, as shown below for the 14th frequency bin, $f\sim 14 f_1$, where $f_1$ is the frequency in the 1st bin.

```python
res2_unit=1e-18
k_index=13
ra2_scaled=(4e-12)*(k_index+1)**(-13/3) # rescale amplitude according to GWB power spectrum
ppkstats=free.ppkstats
fig, ax=plt.subplots(nrows=4,ncols=2,sharex=True, sharey='row', figsize=(10,10))
free.plot_rkrk(ax, res2_unit=res2_unit, k_index=k_index, \
               color='blue', fmt='o', markersize=5, elinewidth=1, capsize=5)
free.plot_rkrk_model(ax, ra2=ra2_scaled, res2_unit=res2_unit, color='blue', ls='-')

soft.plot_rkrk(ax, res2_unit=res2_unit, k_index=k_index, \
               color='green', fmt='*', markersize=5, elinewidth=1, capsize=3)
mild.plot_rkrk(ax, res2_unit=res2_unit, k_index=k_index, \
               color='red', fmt='s', markersize=5, elinewidth=1, capsize=1)
# loud.plot_rkrk(ax, res2_unit=res2_unit, k_index=k_index, \
#                color='black', fmt='^', markersize=1, elinewidth=1, capsize=0)
    
ax[0,0].set_ylabel(r'${\cal E}_1$ (ns$^2$)')
ax[1,0].set_ylabel(r'${\cal V}_2$ (ns$^4$)')
ax[2,0].set_ylabel(r'${\cal S}_3$ (ns$^6$)')
ax[3,0].set_ylabel(r'${\cal K}_4$ (ns$^8$)')
ax[0,0].set_title(rf'$\alpha_{{(a{k_index+1})}} \alpha_{{(b{k_index+1})}}$')
ax[0,1].set_title(rf'$\beta_{{(a{k_index+1})}} \beta_{{(b{k_index+1})}}$')

ax[0,0].set_yscale('symlog'); ax[0,1].set_yscale('symlog')
ax[1,0].set_yscale('log'); ax[1,1].set_yscale('log')
ax[2,0].set_yscale('symlog'); ax[2,1].set_yscale('symlog')
ax[3,0].set_yscale('log'); ax[3,1].set_yscale('log')

ax[2,0].set_yticks([-1e5, 0, 1e5]); ax[2,1].set_yticks([-1e5, 0, 1e5])

ax[3,0].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,1].set_xlabel(r'Pulsar-pair angular separation, $\zeta$')
ax[3,0].set_xticks(np.arange(0,180,20))
ax[3,0].set_xlim(0, 180); ax[3,1].set_xlim(0, 180)
# move ticks to right for right column axes
for row in range(len(ax)):
    ax[row, 1].yaxis.tick_right()
    ax[row, 1].tick_params(labelright=True, labelleft=False)
fig.tight_layout()
fig.subplots_adjust(hspace=0,wspace=0)
plt.show()
```

#### 2 Density plots at various angles

It is worth looking at the histograms at fixed frequency bins and angular separation in order to see the actual distribution of the samples that would not necessarily be captured by mere error bars. This is shown below for the mean in several angles in the noise-free case in the 1st and 14th frequency bins; this is the actual ensemble distribution of the HD correlation.

```python
# 1st frequency bin
zinds=[0,2,12,13] # angular separation indices to plot
fig, ax = plt.subplots(figsize=(5,5))
k_index=0
free.plot_rkrk_density(ax=ax, stats_key='E1', ensemble_stats_key='akak_ensemble_stats', \
                       k_index=k_index, res2_unit=1e-12, zinds=zinds)
ax.set_xlabel(rf'$\langle \alpha_{{a{k_index+1}}} \alpha_{{b{k_index+1}}} \rangle$ ($\mu$s$^2$)')
ax.set_ylabel('Density')
ax.legend(loc='upper right')
fig.tight_layout()
plt.show()

# 14th frequency bin
fig, ax = plt.subplots(figsize=(5,5))
k_index=13
free.plot_rkrk_density(ax=ax, stats_key='E1', ensemble_stats_key='akak_ensemble_stats', \
                       k_index=k_index, res2_unit=1e-18, zinds=zinds)
ax.set_xlabel(rf'$\langle \alpha_{{a{k_index+1}}} \alpha_{{b{k_index+1}}} \rangle$ (ns$^2$)')
ax.set_ylabel('Density')
ax.legend(loc='upper right')
fig.tight_layout()
plt.show()
```

The use of another frequency bin is of course unnecessary, but it's a good check nonetheless. The similar distributions can also be checked in the sample pair statistics of $\beta_{ak}$'s. Mainly, the spatial correlation that is brought by the GWB tilts the ensemble distributions away from a Gaussian distribution (Appendix A in [2407.17987](https://arxiv.org/abs/2407.17987)). The distribution of the higher cumulants can also easily be checked, as follows for the variance and the skewness.

```python
# 2pf variance in the 1st frequency bin
zinds=[0,2,7,8,9,12,13] # angular separation indices to plot
fig, ax = plt.subplots(figsize=(10,5), ncols=2)
free.plot_rkrk_density(ax=ax[0], stats_key='V2', ensemble_stats_key='akak_ensemble_stats', \
                       k_index=0, res2_unit=1e-12, zinds=zinds)
free.plot_rkrk_density(ax=ax[1], stats_key='S3', ensemble_stats_key='akak_ensemble_stats', \
                       k_index=0, res2_unit=1e-12, zinds=zinds)
ax[0].set_xlabel(rf'${{\cal V}}_2 [ \alpha_{{a{k_index+1}}} \alpha_{{b{k_index+1}}} ]$ ($\mu$s$^2$)')
ax[1].set_xlabel(rf'${{\cal S}}_3 [ \alpha_{{a{k_index+1}}} \alpha_{{b{k_index+1}}} ]$ ($\mu$s$^2$)')
ax[0].set_ylabel('Density')
ax[0].legend(loc='upper right')

ax[1].set_yscale('log')
ax[1].yaxis.tick_right()
ax[1].tick_params(labelright=True, labelleft=False)
fig.tight_layout()
fig.subplots_adjust(wspace=0)
plt.show()
```
