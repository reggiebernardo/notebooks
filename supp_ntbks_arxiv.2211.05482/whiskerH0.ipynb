```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(10)
```

```python
# Creating dataset

# ABC: LCDM
lcdm_sc = np.random.normal(67.8, 0.1, 100000)
lcdm_scb = np.random.normal(67.0, 0.5, 100000)
lcdm_scbh1 = np.random.normal(67.8, 0.3, 100000)
lcdm_scbh2 = np.random.normal(68.5, 0.2, 100000)
# lcdm_data = [lcdm_sc, lcdm_scb, lcdm_scbh1, lcdm_scbh2]
lcdm_data = [lcdm_scbh2, lcdm_scbh1, lcdm_scb, lcdm_sc]

# ABC: X2CDM
x2cdm_sc = np.random.normal(67.87, 0.03, 100000)
x2cdm_scb = np.random.normal(67.1, 0.2, 100000)
x2cdm_scbh1 = np.random.normal(67.2, 0.2, 100000)
x2cdm_scbh2 = np.random.normal(67.2, 0.1, 100000)
x2cdm_data = [x2cdm_scbh2, x2cdm_scbh1, x2cdm_scb, x2cdm_sc]

# ABC: X3CDM
x3cdm_sc = np.random.normal(67.92, 0.03, 100000)
x3cdm_scb = np.random.normal(67.97, 0.02, 100000)
x3cdm_scbh1 = np.random.normal(67.97, 0.01, 100000)
x3cdm_scbh2 = np.random.normal(67.9, 0.2, 100000)
x3cdm_data = [x3cdm_scbh2, x3cdm_scbh1, x3cdm_scb, x3cdm_sc]

data_abc = lcdm_data + x2cdm_data + x3cdm_data

# MCMC: LCDM
mcmc_lcdm_sc = np.random.normal(67.4, 1.6, 100000)
mcmc_lcdm_scb = np.random.normal(67.9, 0.8, 100000)
mcmc_lcdm_scbh1 = np.random.normal(67.5, 0.4, 100000)
mcmc_lcdm_scbh2 = np.random.normal(67.0, 0.6, 100000)
mcmc_lcdm_data = [mcmc_lcdm_scbh2, mcmc_lcdm_scbh1, 
                  mcmc_lcdm_scb, mcmc_lcdm_sc]
 
# MCMC: X2CDM
mcmc_x2cdm_sc = np.random.normal(67.7, 1.8, 100000)
mcmc_x2cdm_scb = np.random.normal(67.0, 0.8, 100000)
mcmc_x2cdm_scbh1 = np.random.normal(67.3, 0.4, 100000)
mcmc_x2cdm_scbh2 = np.random.normal(69.6, 0.7, 100000)
mcmc_x2cdm_data = [mcmc_x2cdm_scbh2, mcmc_x2cdm_scbh1, 
                   mcmc_x2cdm_scb, mcmc_x2cdm_sc]

# MCMC: X3CDM
mcmc_x3cdm_sc = np.random.normal(67.7, 1.8, 100000)
mcmc_x3cdm_scb = np.random.normal(66.9, 0.8, 100000)
mcmc_x3cdm_scbh1 = np.random.normal(67.2, 0.4, 100000)
mcmc_x3cdm_scbh2 = np.random.normal(69.6, 0.7, 100000)
mcmc_x3cdm_data = [mcmc_x3cdm_scbh2, mcmc_x3cdm_scbh1, 
                   mcmc_x3cdm_scb, mcmc_x3cdm_sc]

data_mcmc = mcmc_lcdm_data + mcmc_x2cdm_data + mcmc_x3cdm_data

data = data_mcmc + data_abc

fig = plt.figure(figsize =(10, 10))
ax = fig.add_subplot(111)
 
# Creating axes instance
# bp = ax.boxplot(data, patch_artist = True,
#                 notch ='True', vert = 0)

bp = ax.boxplot(data, notch = True, patch_artist = True, \
                vert = 0, whis = 0.5, \
                showbox = True, showfliers = False, showmeans = False)
 
colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF',
          '#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']
 
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
 
# changing color and linewidth of
# whiskers
for whisker in bp['whiskers']:
    whisker.set(color ='#8B008B',
                linewidth = 1.5,
                linestyle ="-")
 
# changing color and linewidth of
# caps
for cap in bp['caps']:
    cap.set(color ='#8B008B',
            linewidth = 2)
 
# changing color and linewidth of
# medians
for median in bp['medians']:
    median.set(color ='red',
               linewidth = 2)
 
# changing style of fliers
# for flier in bp['fliers']:
#     flier.set(marker ='D',
#               color ='#e7298a',
#               alpha = 0.5)
     
# x-axis labels
ax.set_yticklabels([r'$\Lambda$CDM: SNe + CC + BAO + $H_0^{\rm R22}$', # MCMC
                    r'$\Lambda$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$\Lambda$CDM: SNe + CC + BAO', 
                    r'$\Lambda$CDM: SNe + CC',
                    r'$X_2$CDM: SNe + CC + BAO + $H_0^{\rm R22}$',
                    r'$X_2$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$X_2$CDM: SNe + CC + BAO', 
                    r'$X_2$CDM: SNe + CC',
                    r'$X_3$CDM: SNe + CC + BAO + $H_0^{\rm R22}$',
                    r'$X_3$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$X_3$CDM: SNe + CC + BAO', 
                    r'$X_3$CDM: SNe + CC',
                    r'$\Lambda$CDM: SNe + CC + BAO + $H_0^{\rm R22}$', # ABC
                    r'$\Lambda$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$\Lambda$CDM: SNe + CC + BAO', 
                    r'$\Lambda$CDM: SNe + CC',
                    r'$X_2$CDM: SNe + CC + BAO + $H_0^{\rm R22}$',
                    r'$X_2$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$X_2$CDM: SNe + CC + BAO', 
                    r'$X_2$CDM: SNe + CC',
                    r'$X_3$CDM: SNe + CC + BAO + $H_0^{\rm R22}$',
                    r'$X_3$CDM: SNe + CC + BAO + $H_0^{\rm P18}$',
                    r'$X_3$CDM: SNe + CC + BAO', 
                    r'$X_3$CDM: SNe + CC'])
 
# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

# H0 references

### Plot the vertical bars for reference: R20 vs CMB
ax.bar(73.30, 100, width = 2*1.04, facecolor = 'cyan', alpha = 0.5)
ax.bar(73.30, 100, width = 4*1.04, facecolor = 'cyan', alpha = 0.2)
ax.bar(67.4, 100, width = 2*0.5, facecolor = 'pink', alpha = 0.8)
ax.bar(67.4, 100, width = 4*0.5, facecolor = 'pink', alpha = 0.4)
ax.set_ylim(top = 25)

ax.set_xlabel('$H_0$ [km s$^{-1}$ Mpc$^{-1}$]')

# horizontal line reference
ax.axhline(12.5, c = 'k', linewidth = 3)
ax.text(70, 11.0, 'MCMC', rotation = 0, fontsize = 23)
ax.text(70, 13.0, 'ABC', rotation = 0, fontsize = 23)

# show plot
fig.savefig('figs/whiskerH0.pdf', bbox_inches = 'tight')
plt.show()
```
